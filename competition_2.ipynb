# Stacking

```python
from sklearn.ensemble import StackingRegressor
```

```python
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LinearRegression
from catboost import CatBoostRegressor
```

```python
estimators = [('svr', SVR()),
              ('lr', RidgeCV()),
                            #('rf', RandomForestRegressor(n_estimators=1100, max_depth=10, n_jobs=-1)),
              ('lgbm', LGBMRegressor(num_leaves=128, learning_rate=0.01, n_estimators=900)),
              ('knn', KNeighborsRegressor()),
              ('cb', CatBoostRegressor(n_estimators=1000, max_depth=10, learning_rate=0.01))]

stack = StackingRegressor(estimators=estimators, 
                          final_estimator=RandomForestRegressor(n_estimators=1000, max_depth=10), cv=5, n_jobs=-1)
stack.fit(Xtrain, Ytrain)
```

```python
pred = stack.predict(Xtest)
mean_squared_log_error(Ytest, pred)
```
